```python
import pandas as pd
import yfinance as yf
gold_data = pd.read_csv("goldprices.csv")
```

```python
tickers = {
    "Gold_YF": "GC=F",
    "DXY": "DX-Y.NYB",
    "TNX": "^TNX",
    "VIX": "^VIX",
    "CrdOil": "CL=F",
    "Cu": "HG=F",
    "Ag": "SI=F",
    "NatGas": "NG=F",
    "Corn": "ZC=F",
    "Soya": "ZS=F",
    "Cattle": "LE=F",
    "Coco": "CC=F",
    "Coffee": "KC=F",
    "Cotton": "CT=F"    
}

data = {}

for name, ticker in tickers.items():
    df = yf.download(ticker, start="2000-08-30", interval="1d") #2000-08-30 is where gold YF starts from
    df = df[["Close"]]
    df.rename(columns={"Close": name}, inplace=True)
    data[name] = df
```

```python
merged_df = pd.concat(data.values(), axis=1)
merged_df.columns = merged_df.columns.get_level_values(0)
```

```python
merged_df.reset_index(inplace=True) 
merged_df.columns.name = "Index"
```

```python
# Calculating the Exceptional VIX values ranging above the std deviation where 0 = mean, -1 = -1*sigma, +1 = 1*sigma and so on...
mean_vix = merged_df['VIX'].mean()
std_vix = merged_df['VIX'].std()

merged_df['VIX_PCE'] = ((merged_df['VIX'] - mean_vix) / std_vix)

merged_df
```

```python
merged_df.to_csv("gold_macro_yfinance_2000.csv")
```

### We observe that there are some rows where gold_yf populates but DXY, TNX, VIX don't, we neeed to forward fill these values so that the last known value is still in place.,

```python
merged_df.fillna(method="ffill", inplace=True)
```

```python
merged_df["Gold_return_1d"] = merged_df["Gold_YF"].pct_change()
```

```python
merged_df["Gold_YF"].shift(5)[:10]
```

```python
merged_df["Gold_forward_5d"] = ((merged_df["Gold_YF"].shift(-5) - merged_df["Gold_YF"])/merged_df["Gold_YF"])
```

```python
merged_df
```

```python
import numpy as np
prices = merged_df["Gold_YF"].values
forward_5d = (prices[5:] - prices[:-5]) / prices[:-5]
merged_df["new_test_5dF"] = np.nan
```

```python
merged_df.loc[merged_df.index[:len(forward_5d)], "new_test_5dF"] = forward_5d
```

```python
merged_df[5500:6000]
```

```python
def target(x):
    if x >= 0.02:
        return 1
    elif x <= -0.02:
        return -1
    else:
        return 0
```

```python
merged_df["Target"] = merged_df["Gold_forward_5d"].apply(target)
```

```python
merged_df["Target"].value_counts(normalize=True)
```

```python
# Easier Method
merged_df["Target"] = 0
merged_df.loc[merged_df["Gold_forward_5d"] >= 0.02, "Target"] = 1
merged_df.loc[merged_df["Gold_forward_5d"] <= -0.02, "Target"] = -1
```
